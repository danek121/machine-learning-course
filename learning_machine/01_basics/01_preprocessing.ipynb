```python
#import bibliotek numpy, pandas oraz sklearn
import numpy as np
import pandas as pd
import sklearn

#wypisanie wersji biblioteki sklearn
sklearn.__version__
```

```python
#generowanie danych
#utworzenie obiekt√≥w dateframe gdzie kluczem sa nazwy kolumn
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

# tworzenie zmiennej i utworzenie obiektu dateframe

df_raw = pd.DataFrame(data=data)
df_raw


```

```python
#tworzenie kopii danych oraz wyswietlenie informacji o obiekcie
df = df_raw.copy()
df.info()
```

```python
#zmiana typu danych
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

  df['weight'] = df['weight'].astype('float')

  df.info()
```

```python
#wyswietlenie podstawowych statystyk obiektu dateframe
#describe wyswietla tylko numeryczne kolumny
df.describe()
```

```python
#transponujemy kolumny
df.describe().T
```

```python
#dodajemy wyswietlanie statystyki ze zmiennymi kategorycznymi

df.describe(include=['category']).T
```

```python
#wyswietlanie obiektu df
df
```

```python
#przygotowanie danych
#le.fit_transform zwraca tablice numeryczna(number array)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit_transform(df['bought'])
```

```python
#przypisanie zmienionych danych do (no = 0;yes=1)
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
#zmiana do pierwotnego stanu
df['bought'] = le.inverse_transform(df['bought'])
df
```

```python
#importowanie onehotencoder(biblioteka do pobierania danych) z sklearn preprocessing
#utworzenie instancje klasy

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
encoder.fit(df[['size']])
```

```python
#wykonanie kodowania uzywajac transform
encoder.transform(df[['size']])
```

```python
#wyswietlanie kategorii i jak czytac
encoder.categories_
```

```python
#zamiana yes/no na 1/0
encoder = OneHotEncoder(drop="first", sparse_output=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

```python
#Pandas get_dummies() funkcja

pd.get_dummies(data=df)
```

```python
#parametr drop first do usuwania pierwszej kolumny

pd.get_dummies(data=df, drop_first=True)
```

```python
pd.get_dummies(data=df, drop_first=True, prefix='new') #dodaje slowo przed nazwa
pd.get_dummies(data=df, drop_first=True, prefix_sep='-') #dodaje - pomiedzy nazwami
pd.get_dummies(data=df, drop_first=True, columns=['size']) #wybrane kolumny sa prefixowane
```

```python
#standaryzacja danych biblioteka sklearn
# std() pandas nieobciazona
# std() numpy obciazony


print(f"{df['price']}\n")
print(f"≈örednia: {df['price'].mean()}")
print(f"Odchylenie Standardowe: {df['price'].std():.4f}")

```

```python
#wystandaryzowane danych (oscylujemy blisko 0)
 (df['price'] - df['price'].mean()) / df['price'].std()
```

```python
#wystandaryzowane danych (oscylujemy blisko 0)
def standarize(x):
  return (x-x.mean()) / x.std()
standarize(df['price'])
```

```python
#wystandaryzowane danych (oscylujemy blisko 0) funkcja scale z skleanr(inne wartosci bo obciazony)
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
#funkcja standardscale lepsza od poprzedniej bo zapisuje wczesniejsze dane
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])



```

```python
#dodanie wartosci do danych
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```
